## Outlier Analysis

```python
data_pca_df.describe()
```

From the above, `income` seems to have strongly varying values since its Standard Deviation is above 8. Let us find the records whose 

There are two main methods [[1]( https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba)] for outlier analysis and they are:
1. Using Z-score
2. Using the Interquantile distance

Let us use both these methods and then whether we should drop countries

**Reference**
1. https://towardsdatascience.com/ways-to-detect-and-remove-the-outliers-404d16608dba

```python
data_pca_df
```

### Using Z-Score

```python
from scipy import stats
z = np.abs(stats.zscore(data_pca_df[list(data_pca_df.columns[:-1])]))
```

```python
print(np.where(z>3))
```

There are 5 countries whose Z-score is greater than 3 and hence have to be dropped. Let us see the countries:

```python
countries_zscore = set(data_pca_df[(z>3).any(axis=1)]['country'])
countries_zscore
```

### Using Inter-Quantile Range

```python
q1 = data_pca_df.quantile(0.25)
q3 = data_pca_df.quantile(0.75)
IQR = q3 - q1
```

```python
countries_iqr = set(data_pca_df[((data_pca_df<(q1 - 1.5*IQR))|(data_pca_df>(q3 + 1.5*IQR))).any(axis=1)]['country'])
countries_iqr
```

According to the Inter-Quantile Rating, the above countries have to be removed from analysis

```python
countries_zscore.intersection(countries_iqr)
```

An intersection of the above two, shows that the countries represented by Z-score is a sub-set of the IQR set. Hence, let us drop all the countries represented in the Z-score set.

```python
data_df = data_pca_df[(z<3).all(axis=1)]
data_df
```

Let us create one dataframe where all the faulty countries are dropped and see how that behaves

```python
data_df_dropped = data_pca_df[~((data_pca_df<(q1 - 1.5*IQR))|(data_pca_df>(q3 + 1.5*IQR))).any(axis=1)]
data_df_dropped
```

Let us plot some scatter plots for both the Dataframes and visualise their correlations

## Clustering
Let us begin by verifying if Clustering can be done on the above dataset. For this, we use the `Hopkins Measure`

```python
from sklearn.neighbors import NearestNeighbors
from random import sample
from numpy.random import uniform
import numpy as np
from math import isnan
 
def hopkins(X):
    d = X.shape[1]
    #d = len(vars) # columns
    n = len(X) # rows
    m = int(0.1 * n) 
    nbrs = NearestNeighbors(n_neighbors=1).fit(X.values)
 
    rand_X = sample(range(0, n, 1), m)
 
    ujd = []
    wjd = []
    for j in range(0, m):
        u_dist, _ = nbrs.kneighbors(uniform(np.amin(X,axis=0),np.amax(X,axis=0),d).reshape(1, -1), 2, return_distance=True)
        ujd.append(u_dist[0][1])
        w_dist, _ = nbrs.kneighbors(X.iloc[rand_X[j]].values.reshape(1, -1), 2, return_distance=True)
        wjd.append(w_dist[0][1])
 
    H = sum(ujd) / (sum(ujd) + sum(wjd))
    if isnan(H):
        print(ujd, wjd)
        H = 0
 
    return H
```

```python
hopkins(data_df[list(data_df.columns[:-1])])
```

```python
hopkins(data_df_dropped[list(data_df_dropped.columns[:-1])])
```

From the above values, the dataFrame `data_df` has a higher tendency to cluster. This is enough indication that we have to retain the other countries except the ones that were dropped in the `intersection` step. Since the Hopkin measure is greater than 7, the dataset is suitable for clustering using K-Means Clustering

### K-Means Clustering

Let us pop the countries out before we proceed into clustering

```python
countries = data_df.pop('country')
```

#### Optimal Clusters
Before we create the model, lets find the optimal clusters for the current dataset. We can do this by using the following 
methods:

    -Elbow Curve
    -Silhouette score
Lets start off by setting a range of potential clusters. Since, we are dealing with the Indian Continent and its fight to eradicate poverty, there is a good possibility that there could be a large number of clusters forming which face similar problems

```python
from sklearn.cluster import KMeans
cluster_ranges = list(range(2,20))
scores = []
for val in cluster_ranges:
    kmeans = KMeans(n_clusters=val,max_iter=50,random_state=123).fit(data_df)
    scores.append(kmeans.inertia_)
plt.figure(figsize=(16,9))
sns.lineplot(x=cluster_ranges,y=scores)    
```

There seems to be a nominal drop after 8, lets plot the first 8 values and see how the distance reduces

```python
sns.lineplot(x=cluster_ranges[:8],y=scores[:8]) 
```

From the above figure, there is a drastic fall in distance when we increase the clusters from 2 to 3. From 3 to 4 there is a smaller, yet perceivable drop. From 6 to 9 the graph is close to a linear descent. Hence, our ideal clusters should be 3 or 4. Let us confirm the same using the Silhouette cover method

```python
from sklearn.metrics import silhouette_score
cluster_ranges = list(range(2,9))
scores = []
for value in cluster_ranges:
    kmeans = KMeans(n_clusters=value,max_iter=50,random_state=123).fit(data_df)
    cluster_labels = kmeans.labels_
    silhouette_avg  = silhouette_score(data_df, cluster_labels)
    print("For no of clusters = {}, the silhoutte_score is {}".format(value, silhouette_avg))
```

Between 3 and 4, `3` has the higher Silhoutte score, hence we should go with 3 clusters. An interesting observation here: 5 clusters, though less than 3, has a higher silhoutte score than 4. Hence, if we were to pick clusters, the order would be as follows:

**3>5>4>2**

Lets create our final clustering model and join the result with the dataset

```python
data_df_modeling = data_df.copy()
```

```python
kmeans_final = KMeans(n_clusters=3,max_iter=50,random_state=123).fit(data_df)
data_df_modeling['clusterid'] = kmeans_final.labels_
data_df_modeling['country'] = countries
```

```python
data_df_modeling.groupby(['clusterid']).nunique().plot(kind='bar',figsize=(16,9))
```

## Hierarchical Clustering

```python
from scipy.cluster.hierarchy import linkage,dendrogram,cut_tree

#single-linkage
mergings_single = linkage(data_df,method='single',metric='euclidean')
dendrogram(mergings_single)
plt.show()
```

```python
#complete-linkage
mergings_complete = linkage(data_df,method='complete',metric='euclidean')
dendrogram(mergings_complete)
plt.show()
```

```python
#Lets create 3 clusters for single linkage
data_hier_single3 = data_df.copy()
data_hier_single3['clusterid'] = cut_tree(mergings_single, n_clusters=3).reshape(-1, )
data_hier_single3['country'] = countries

#Lets create 5 clusters for single linkage
data_hier_single5 = data_df.copy()
data_hier_single5['clusterid'] = cut_tree(mergings_single, n_clusters=5).reshape(-1, )
data_hier_single5['country'] = countries
```

```python
#Lets create 3 clusters for complete linkage
data_hier_comp3 = data_df.copy()
data_hier_comp3['clusterid'] = cut_tree(mergings_complete, n_clusters=3).reshape(-1, )
data_hier_comp3['country'] = countries

#Lets create 5 clusters for complete linkage
data_hier_comp5 = data_df.copy()
data_hier_comp5['clusterid'] = cut_tree(mergings_complete, n_clusters=5).reshape(-1, )
data_hier_comp5['country'] = countries
```
